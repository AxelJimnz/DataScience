# Carga de datos:
- Utiliza Pandas para cargar el conjunto de datos.



```python
import pandas as pd
import os
```

- Creamos una variable de datos donde se guardará el archivo Excel llamada "data"

```python
ruta_archivo_excel = 'datos/AD_CL.xlsx'
data = pd.read_excel(ruta_archivo_excel)
```

# Exploración de datos:

- Muestra las primeras filas.
- Obtiene estadísticas descriptivas.
- Verifica valores nulos.

```python
print(data.head())
print(data.describe())
print(data.isnull().sum())
```

# Limpieza de datos:

- Elimina filas con valores nulos.
- Llena valores nulos con la media.
- Elimina duplicados si es necesario.

```python
data = data.dropna()
data.fillna(data.mean(), inplace=True)
data = data.drop_duplicates()
```

```python
print(data)
```
